### subway_semi_final 데이터 셋에 역 주변 시설 정보를 결합
- **최종 데이터 셋인 subway_final 데이터 셋을 생성!!**

```python
import pandas as pd
import numpy as np
```

**subway_semi_final 데이터 셋 불러오기**

```python
subway_semi_final = pd.read_csv("C:/Users/Playdata/Playdata_anl/1st_Project/dataset/subway/subway_semi_final.csv")
subway_semi_final.drop(['Unnamed: 0'], axis = 1, inplace = True) # 첫 번째 컬럼 제거(의미 없는 컬럼임)
subway_semi_final.head()
```

- **지하철 데이터 셋의 dong 변수**에는 "**법정동**"이 들어가있다.
- **인구 데이터 셋의 dong 변수**에는 "**행정동**"이 들어가있다.
    - 따라서 **인구 데이터 셋의 "행정동"을 "법정동"으로 변경해주는 작업이 필요**했었다.
    - 이 부분은 다른 팀원 분들께서 **"지하철 역 주소의 행정동"**을 직접 입력한 데이터 셋을 만듦으로써 해결되었다!!

- 먼저 만들어진 "행정동 데이터 셋"을 "suway_semi_final 데이터 셋"에 **inner join** 해주겠다.
    - 결합시키기 전에, subway_semi_final 데이터 셋의 존재하고 있는 **dong(법정동) 변수는 삭제**시켜주겠다.
        - 그리고 **subway_semi_final 데이터 셋**에서 **transfer, transfer_num, entrance, gate 변수들은 제거**해주겠다.
        - 지훈님이 새로 만들어주신 subway_info_total_final 데이터 셋에 **업데이트 되었을 수도 있기 때문에!!**
    - **station_name 변수를 조인키로 사용**하여 **inner join** 할 것이다.
        - 그러면 subway_semi_final 데이터 셋에는 **dong(행정동) 변수가 새롭게 추가될 것**이다.

```python
# 기존의 dong(법정동), transfer, transfer_num, entrance, gate 변수 제거
subway_semi_final.drop(['dong', 'transfer', 'transfer_num', 'entrance', 'gate'], axis = 1, inplace = True)
```

```python
subway_semi_final.head()
```

```python
subway_info_total_final = pd.read_csv("C:/Users/Playdata/Playdata_anl/1st_Project/dataset/subway/subway_info_total_final.csv")
subway_info_total_final.drop(['Unnamed: 0'], axis = 1, inplace = True) # 첫 번째 컬럼 제거(의미 없는 컬럼임)
subway_info_total_final
```

```python
# station_name 변수를 조인키로 inner join
subway_final = pd.merge(subway_semi_final, subway_info_total_final, how = 'inner', on = ['station_name', 'line'])
subway_final
```

**조인키 생성**
- 총 2개의 새로운 컬럼을 생성해야 한다.
    - "date_hour_자치구", "date_hour_행정동" 이런 식으로..
- 그 전에 **먼저 subway_final 데이터 셋의 'gu' 변수 중, "강남"으로 되어있는 것들을 싹 다 "강남구"라고 변경**해주어야 한다.

```python
subway_join_key_ls = subway_final['gu'].to_list()

for idx, value in enumerate(subway_join_key_ls):
    if value == '강남':
        subway_join_key_ls[idx] = '강남구'
        
df_subway = pd.DataFrame(subway_join_key_ls)
df_subway[df_subway[0] == '강남'] # "강남"인 것이 미존재하는 것이 맞음!!
```

```python
subway_final['gu'] = subway_join_key_ls
subway_final[subway_final['gu']  == '강남'] # "강남"인 것이 미존재하는 것이 맞음!!
```

```python
# 인구 데이터와 결합할 때 사용할 조인키
subway_final['pop_join_key'] = subway_final['date'] + '_' + subway_final['dong']

# 기상 데이터와 결합할 때 사용할 조인키
subway_final['weather_join_key'] = subway_final['date'] + '_' + subway_final['gu']
```

- **address 변수**를 자세히 살펴보니...**"경기도", "인천광역시", "서울툭별시"** 이렇게 분류되어 있음을 확인할 수 있었다. (일 똑바로 안하네..)
    - **"경기도"와 "인천광역시"**는 **나중에 행(row) 자체를 제거**해주고, 우선 **"서울툭별시"**를 **"서울특별시"로 수정**해주겠다.

```python
subway_seoul_add = subway_final['address'].str.split().to_list()
subway_seoul_add
```

```python
result = []
for i in subway_seoul_add:
    if i[0] == '서울툭별시':
        i[0] = '서울특별시'
    result.append(' '.join(i))
    
result
```

```python
result = pd.DataFrame(result)
result[result[0].str.contains('서울툭별시')]
```

- 위 결과에 아무 것도 안 뜨는 것으로 보아, '서울툭별시'가 '서울특별시'로 잘 변경되었음을 알 수 있다.

```python
len(result)
```

```python
subway_final['address'] = result
```

```python
# '서울툭별시' 변경 되었는지 검토용
check1 = subway_final[subway_final['address'].str.contains('서울특별시')]
check2 = subway_final[subway_final['address'].str.contains('경기도')]
check3 = subway_final[subway_final['address'].str.contains('인천광역시')]

len(check1) + len(check2) + len(check3)
```
