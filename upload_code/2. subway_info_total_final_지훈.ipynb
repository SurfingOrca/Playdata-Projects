```python
import pandas as pd
```

# 1. 지하철 전체 정보

```python
# sit = station_info_total
sit = pd.read_csv('data/subway_info_total.csv')
sit
```

# 2. 역 출구별 주변주요시설 정보

```python
# nsi = near_station_info
nsi = pd.read_csv('data/near_station_info_total.csv')
nsi
```

## 2-1. 역별 주요시설수 카운팅

```python
nsi_cnt = [nsi[(nsi['station_name'] == i) & (nsi['line'] == j)]['line'].count() for i, j in zip(sit['station_name'], sit['line'])]
sum(nsi_cnt)
```

## 2-2. 지하철 전체정보에 포함시키기

```python
sit_b = sit
sit_b['nsi_cnt'] = nsi_cnt
sit_b
```

# 5. 대학교 정보

```python
# uvi = universities_info
uvi = pd.read_csv('data/universities_seoul.csv')
uvi
```

## 5-1. (지하철명, 호선) 튜플을 서로 비교하여 오류값이 존재하는지 확인

```python
# 전체 지하철 정보에서 튜플 만들기 (지하철명, 호선)
temp = list(map(lambda x, y: (x, y), sit_b['station_name'].to_list(), sit_b['line'].to_list()))
len(temp)
```

```python
# 대학교 정보에서 튜플 만들기 (지하철명, 호선)
temp2 = list(map(lambda x, y: (x, y), uvi['지하철명'].to_list(), uvi['호선'].to_list()))
temp2
```

```python
# 오류값이 있으면 temp2에 남도록 구현
for i in temp:
    while i in temp2:
        temp2.pop(temp2.index(i))
temp2
```

## 5-2. 역별 대학교수 카운팅

```python
uvi_cnt = [uvi[(uvi['지하철명'] == i) & (uvi['호선'] == j)]['호선'].count() for i, j in zip(sit['station_name'], sit['line'])]
sum(uvi_cnt)
```

## 5-3. 지하철 전체정보에 포함시키기

```python
sit_b['uvi_cnt'] = uvi_cnt
sit_b
```
