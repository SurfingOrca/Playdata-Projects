# 변환 함수
- "20대, 30대, 40대, 50대, 60대" 변수들에 대해서는 하나의 함수를 만들어서 반복적인 작업을 최소화!!

````python
def trans_pop(age):
    pop = populations[['pop_join_key', 'hour', f'{age}대']]
    
    pop_pivot = pd.pivot_table(pop, index = ['pop_join_key', 'hour'], values = [f'{age}대'])
    pop_pivot = pop_pivot.unstack()
    
    pop_temp = pd.DataFrame(pop_pivot)
    pop_temp[(f'{age}대', 0)].to_list()
    
    pop_final = populations[['pop_join_key']]
    pop_final.drop_duplicates(inplace = True)
    pop_final.reset_index(drop = True, inplace = True)
    
    for i in list(range(5, 24)) + [0]:
        pop_final[f'pop_{age}_{i}'] = pop_temp[(f'{age}대', i)].to_list()
        
    pop_final.to_csv(f'C:/Users/Playdata/Playdata_anl/1st_Project/dataset/population/total_final/pop_{age}_final.csv')
````

````python
for i in range(20, 70, 10):
    trans_pop(i)
````
