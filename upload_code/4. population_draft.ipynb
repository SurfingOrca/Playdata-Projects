### 2017~2019년도의 인구 데이터 결합
- ```glob``` 모듈을 사용하여 여러 개의 csv 파일을 한 번에 불러오기 
- Pandas의 ```concat()``` 함수를 사용해서 pop_total 데이터 셋 생성

````python
from tqdm.notebook import tqdm
import pandas as pd
import glob

path = r'C:/Users/Playdata/Playdata_anl/1st_Project/dataset/population'
all_files = glob.glob(path + "/*.csv")

files = []

for filename in tqdm(all_files):
    pop = pd.read_csv(filename, error_bad_lines = False)
    files.append(pop)

files
````

````python
pop_total = pd.concat(files, axis = 0, ignore_index = True)
pop_total.drop(['Unnamed: 0'], axis = 1, inplace = True) # 첫 번째 컬럼 제거(의미 없는 컬럼임)
````

- pop_total 데이터 셋의 변수명을 영어로 변경

````python
new_colnames = {'기준일ID':'date', '시간대구분':'hour', '행정동코드':'dong_code', '총생활인구수':'total', 
                '남자0세부터9세생활인구수':'m_0_9', '남자10세부터14세생활인구수':'m_10_14', '남자15세부터19세생활인구수':'m_15_19',
                '남자20세부터24세생활인구수':'m_20_24', '남자25세부터29세생활인구수':'m_25_29', '남자30세부터34세생활인구수':'m_30_34',
                '남자35세부터39세생활인구수':'m_35_39', '남자40세부터44세생활인구수':'m_40_44', '남자45세부터49세생활인구수':'m_45_49',
                '남자50세부터54세생활인구수':'m_50_54', '남자55세부터59세생활인구수':'m_55_59', '남자60세부터64세생활인구수':'m_60_64',
                '남자65세부터69세생활인구수':'m_65_69', '남자70세이상생활인구수':'m_over_70', '여자0세부터9세생활인구수':'f_0_9',
                '여자10세부터14세생활인구수':'f_10_14', '여자15세부터19세생활인구수':'f_15_19', '여자20세부터24세생활인구수':'f_20_24',
                '여자25세부터29세생활인구수':'f_25_29', '여자30세부터34세생활인구수':'f_30_34', '여자35세부터39세생활인구수':'f_35_39',
                '여자40세부터44세생활인구수':'f_40_44', '여자45세부터49세생활인구수':'f_45_49', '여자50세부터54세생활인구수':'f_50_54',
                '여자55세부터59세생활인구수':'f_55_59', '여자60세부터64세생활인구수':'f_60_64', '여자65세부터69세생활인구수':'f_65_69',
                '여자70세이상생활인구수':'f_over_70'}

pop_total.rename(columns = new_colnames, inplace = True)
````

- pop_total 데이터 셋을 남/여 구분 없이 연령대 별로 합쳐주기
    - 총 7개의 그룹으로 범주화한 연령대 변수 생성
        - 20세미만
        - 20대
        - 30대
        - 40대
        - 50대
        - 60대
        - 70세이상

````python
# 총 7개의 그룹으로 범주화한 연령대 변수 생성
pop_total['20세미만'] = pop_total['m_0_9'] + pop_total['m_10_14'] + pop_total['m_15_19'] + pop_total['f_0_9'] + pop_total['f_10_14'] + pop_total['f_15_19']
pop_total['20대'] = pop_total['m_20_24'] + pop_total['m_25_29'] + pop_total['f_20_24'] + pop_total['f_25_29']
pop_total['30대'] = pop_total['m_30_34'] + pop_total['m_35_39'] + pop_total['f_30_34'] + pop_total['f_35_39']
pop_total['40대'] = pop_total['m_40_44'] + pop_total['m_45_49'] + pop_total['f_40_44'] + pop_total['f_45_49']
pop_total['50대'] = pop_total['m_50_54'] + pop_total['m_55_59'] + pop_total['f_50_54'] + pop_total['f_55_59']
pop_total['60대'] = pop_total['m_60_64'] + pop_total['m_65_69'] + pop_total['f_60_64'] + pop_total['f_65_69']
pop_total['70세이상'] = pop_total['m_over_70'] + pop_total['f_over_70']
    
# 불필요한 변수들 제거
pop_total_final = pop_total.drop(['m_0_9', 'm_10_14', 'm_15_19', 'm_20_24', 'm_25_29', 'm_30_34', 'm_35_39', 
                                  'm_40_44', 'm_45_49','m_50_54', 'm_55_59', 'm_60_64', 'm_65_69', 'm_over_70',
                                  'f_0_9', 'f_10_14', 'f_15_19', 'f_20_24', 'f_25_29', 'f_30_34', 'f_35_39', 
                                  'f_40_44', 'f_45_49','f_50_54', 'f_55_59', 'f_60_64', 'f_65_69', 'f_over_70'], 
                                  axis = 1, inplace = False)

pop_total_final.head()
````

````python
# date 변수의 타입을 str 으로 변경
# <참고> str()로 사용해서 변경하면 이상하게 변경된다...
pop_total_final['date'] = pop_total_final['date'].astype(str)
pop_total_final['date']
````
