```python
# 필요한 라이브러리 import
import pandas as pd
import numpy as np
import datetime
```

# 날씨 data 불러오기

```python
final_weather = pd.read_csv('final_weather(3).csv', encoding = 'cp949')
final_weather
```

```python
# 'Unnamed: 0' column 삭제
final_weather.drop('Unnamed: 0', axis=1, inplace=True)
final_weather
```

# 날씨 8개 특성별 변수 생성 및 행렬 전치 작업

# 날씨 특성 중 '기온'만 남긴 test 변수 생성

```python
test_tem = final_weather[['지점명','주소','구','날짜','시간','기온']].copy()
test_tem
```

```python
test_tem = test_tem.reset_index()
test_tem
```

## pivot_table, unstack 활용해서 시간 특성 컬럼으로 보내기

```python
test_tem_pivot = pd.pivot_table(test_tem, index = ['index','지점명','날짜','시간'], values = ['기온'])   # 셀안의 값 = 기온
test_tem_pivot
```

```python
test_tem_pivot = test_tem_pivot.unstack()
test_tem_pivot
```

```python
test_tem_pivot.head(30)
```

## 데이터 담을 파일 load - 미리 형식 만들어 놓았음

```python
sample = pd.read_csv('sample_form.csv', encoding = 'cp949')
sample
```

```python
# Unnamed: 0 컬럼 삭제
sample.drop('Unnamed: 0', axis =1, inplace = True)
sample
```

## sample index랑 매칭해서 넣어줄 나머지 data

```python
test_tem
```

```python
test_tem2 = test_tem.drop(['index','시간','기온'], axis=1).copy()
test_tem2
```

```python
# 중복값 조회
test_tem2.duplicated().sum()
```

```python
# 중복값삭제
test_tem2 = test_tem2.drop_duplicates()
test_tem2
```

```python
# 27375
```

```python
test_tem2 = test_tem2.reset_index()
test_tem2
```

## 반복문 돌려서 피벗 대각선 값만 담은 list 생성
-  diag 함수 활용해서 list에 담기

```python
t = []
for i in test_tem2['index']:
    t.append(np.diag(test_tem_pivot.iloc[i:i+24]))
```

```python
len(t)
```

```python
t
```

## serise로 변경, sample data set에 하나씩 행추가

```python
for i in range(len(t)):
    sample = sample.append(pd.Series(t[i], index=sample.columns), ignore_index=True)
```

```python
sample # 27375행이여야함
```

## 나머지 정보와 merge (test_tem2, sample)

```python
test_tem2
```

```python
final_tem = pd.merge(test_tem2, sample, how = 'left', on = test_tem2.index)
final_tem
```

```python
# key_0, index 컬럼 삭제
final_tem.drop(['key_0','index'], axis =1, inplace = True)
final_tem
```

```python
# 기온 특성 flie 저장
final_tem.to_csv('final_temperature.csv', encoding='cp949')
```
