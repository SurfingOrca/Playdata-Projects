- 2020-07-10 10:00:00 기준
- 서울시 구별 확진자 수작업으로 입력

```python
covid2 = pd.read_csv('covid2_좌표포함.csv')
covid2
```

## 구별 확진자 시각화를 위해, GeoJSON file load
- 서울시 구별 경계선을 그릴 수 있는 json 파일 얻기
    - 파일 출처: [Github e9t](https://github.com/southkorea/southkorea-maps)
- 'skorea_municipalities_geo_simple.json'

```python
import json

geo_path = 'skorea_municipalities_geo_simple.json'
geo_json = json.load(open(geo_path, encoding = 'utf-8'))
```

## CircleMarker
- 이디야, 범죄현황 분석 참고

```python
covid2.loc[covid2['구'] == '관악구', '확진자수']
```

```python
count = covid2.loc[covid2['구'] == '관악구', '확진자수'][0]
count
```

```python
covid2['확진자수'].sum()
```

```python
covid2['확진자수'] = covid2['확진자수'].astype(float)
```

```python
# 서울시 중심부의 위도, 경도 입니다. 
center = [37.541, 126.986] 

# 맵이 center 에 위치하고, zoom 레벨은 11로 시작하는 맵 m을 만듭니다. 
map = folium.Map(location=center, zoom_start=11) 

# Choropleth 레이어를 만들고, 맵 m에 추가합니다. 
folium.Choropleth( 
    geo_data=geo_json, 
     data=covid2,
     columns=('구', '확진자수'), 
     key_on='feature.properties.name', 
     fill_color='RdPu',    #'RdPu', 'PuRd'
     legend_name='코로나 확진자수'
).add_to(map) # 맵 m을 출력합니다. 

# 들여쓰기 주의!!! for, if, folium 연결되니까!###############
# 두 매장별 매장수, popup message, color
for i in covid2.index:
    size = (covid2.loc[i, '확진자수']/1318*500)
    count = covid2.loc[i, '확진자수']
    gu = covid2.loc[i, '구']
    msg = f'{gu}: {count}명'   #popup message
        
    # 원 크기가 매장수 반영하도록 설정
    folium.CircleMarker(
        [covid2.loc[covid2['구'] == gu,'위도(lat)'],
        covid2.loc[covid2['구'] == gu, '경도(lng)']],
        radius = size,
        color='#FC414E',
        fill_color='#FC414E',
        popup = msg,
        ).add_to(map)

map
```
