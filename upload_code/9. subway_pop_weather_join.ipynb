- 다시 한 번 두 개의 데이터 셋에 gu 변수를 비교해보니, **일치함을 확인**할 수 있다!!

**1. subway_final + weather_final (데이터 셋 결합)**
- **weather_join_key 변수를 조인키로 사용**하여 **left join**
- 두 개의 데이터 셋이 결합된 **subway_weather_final 데이터 셋을 생성!!**

```python
subway_weather_final = pd.merge(subway_final, weather_final, how = 'left', on = 'weather_join_key')
subway_weather_final
```

```python
len(subway_final)
```

- 위 결과를 보면, left join이 잘 이루어진 것을 확인할 수 있다.

```python
subway_weather_final.isnull().sum()
```

- 위의 **41610개의 Null 값들**은 **'경기도', '인천광역시'에 속한 데이터의 개수**이므로 **무시해도 무방**하다!!
    - 이 내용은 코드 맨 마지막 쪽에 자세하게 설명해놓았다.

**2. subway_weather_final + pop_final (데이터 셋 결합)**
- **pop_join_key 변수를 조인키로 사용**하여 **left join**
- 두 개의 데이터 셋이 결합된 **subway_weather_pop_final 데이터 셋을 생성!!**

```python
subway_weather_pop_final = pd.merge(subway_weather_final, pop_final, how = 'left', on = 'pop_join_key')
subway_weather_pop_final
```

```python
len(subway_weather_final)
```

- Left join 후에 존재하는 Null 값들에 대한 고찰...
    - **얼핏 보면 left join이 잘 된 것 같지만, 그래도 다시 한 번 check!!**

```python
cc = subway_weather_pop_final[subway_weather_pop_final['address'].str.contains('서울특별시')]
cc_list = list(set(cc['station_num'])) # 258개
len(cc_list)
```

```python
tt = cc[cc['total_5'].isnull()]
tt_list = list(set(tt['station_num'])) # 256개
len(tt_list)
```

- **"258 - 256 = 2"** 이므로, **차이나는 2개의 지하철 역이 어디인지 파악**해보자.

```python
for i in tt_list:
    if i in cc_list:
        cc_list.pop(cc_list.index(i))
        
cc_list
```

- 확인해보니 **station_num 변수가 2615, 321인 부분에서 차이가 발생**했다.

```python
cc[(cc['station_num'] == 2615) | (cc['station_num'] == 321)]
```
